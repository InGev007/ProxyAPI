```python
import cv2 #імпоруємо opencv
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
```

```python
im = cv2.imread("/content/drive/MyDrive/Colab Notebooks/cat.jpg")
```

```python
im_RGB = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.axis('off') # приховуємо осі
plt.imshow(im_RGB)
plt.show
```

```python
R = im_RGB.copy()
R[:,:,(1,2)] = 0
plt.axis('off')
plt.imshow(R)
plt.show
```

```python
G = im_RGB.copy()
G[:,:,(0,2)] = 0
plt.axis('off')
plt.imshow(G)
plt.show
```

```python
B = im_RGB.copy()
B[:,:,(0,1)] = 0
plt.axis('off')
plt.imshow(B)
plt.show
```

```python
img_HSV = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
plt.axis('off')
plt.imshow(img_HSV)
plt.show
```

```python
H = img_HSV.copy()
# задаємо максимальне (255) значення
# S та V каналам
H[:, :, (0,1)] = 255
# конвертуємо в RGB

plt.show()
plt.axis('off')
plt.imshow(H)
```

```python
H_RGB =  cv2.cvtColor(img_HSV,cv2.COLOR_HSV2RGB)
plt.show()
plt.axis('off')
plt.imshow(H_RGB)
```

```python
V = img_HSV.copy()
# задаємо H максимальне значення(179)
V[:, :, 0] = 179
# задаємо S значення 0
V[:, :, 1] = 0
# конвертуємо в RGB
V_RGB =  cv2.cvtColor(V,cv2.COLOR_HSV2RGB) 
plt.show()
plt.axis('off')
plt.imshow(V_RGB)
```

```python
im_Lab = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
L = im_Lab.copy()
L[:, :, (1,2)] = 0
L_RGB =  cv2.cvtColor(L,cv2.COLOR_LAB2RGB)
L_GRAY = cv2.cvtColor(L_RGB,cv2.COLOR_RGB2GRAY)
# L у чорно-білій гамі кольорів
plt.imshow(L_GRAY, cmap="gray") 
plt.axis('off')
plt.show()
```

```python
a = im_Lab.copy()
a[:, :, 0] = 127
a[:, :, 2] = 0
a_RGB =  cv2.cvtColor(a,cv2.COLOR_LAB2RGB)
a_GRAY = cv2.cvtColor(a_RGB,cv2.COLOR_RGB2GRAY)
```

```python
# створення гами кольорів для відображення
cmap_a = clr.LinearSegmentedColormap.from_list('custom blue', ['Red',"Gray",'Green'], N=255)
plt.imshow(a_GRAY,  cmap=cmap_a)
plt.axis('off')
plt.show()
```
